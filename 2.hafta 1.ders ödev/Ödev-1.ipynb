```python
# Daire alanını hesaplayan fonksiyon
def daire_alani(pi, yaricap):
    alan = pi * yaricap ** 2
    return alan

# Faktöriyel hesaplayan fonksiyon
def faktoriyel(sayi):
    faktoriyel = 1
    for i in range(1, sayi + 1):
        faktoriyel *= i
    return faktoriyel

# Personel sınıfı
class Personel:
    def __init__(self, id, isim, soyisim, dogum_yili, departman, saatlik_ucret, calisma_saati):
        self.id = id
        self.isim = isim
        self.soyisim = soyisim
        self.dogum_yili = dogum_yili
        self.departman = departman
        self.saatlik_ucret = saatlik_ucret
        self.calisma_saati = calisma_saati

    def yas_hesapla(self, yil):
        return yil - self.dogum_yili

    def maas_hesapla(self):
        return self.calisma_saati * self.saatlik_ucret

    def bilgileri_yazdir(self):
        print(f"ID: {self.id}")
        print(f"İsim: {self.isim}")
        print(f"Soyisim: {self.soyisim}")
        print(f"Doğum Yılı: {self.dogum_yili}")
        print(f"Departman: {self.departman}")
        print(f"Haftada Çalışılan Saat: {self.calisma_saati}")
        print(f"Saatlik Ücret: {self.saatlik_ucret}")
        print(f"Yaş: {self.yas_hesapla(2023)}")
        print(f"Maaş: {self.maas_hesapla()}")

# Çalışan sınıfı
class Calisan:
    def __init__(self, ad, soyad, maas):
        self.ad = ad
        self.soyad = soyad
        self.maas = maas

    def calis(self):
        print(f"{self.ad} {self.soyad} işyerinde çalışıyor.")

# Müdür sınıfı
class Mudur(Calisan):
    def __init__(self, ad, soyad, maas, departman, sicil_no):
        super().__init__(ad, soyad, maas)
        self.departman = departman
        self.sicil_no = sicil_no

    def calis(self):
        print(f"{self.ad} {self.soyad}, {self.departman} departmanında müdür olarak çalışıyor. Sicil No: {self.sicil_no}")

# Örnek kullanım
if __name__ == "__main__":
    # Daire alanı hesaplamak için kullanıcıdan bilgi alınır.
    pi = float(input("Pi değerini girin: "))
    yaricap = float(input("Dairenin yarıçapını girin: "))
    alan = daire_alani(pi, yaricap)
    print(f"Daire Alanı: {alan}")

    # Faktöriyel hesaplamak için kullanıcıdan bilgi alınır.
    sayi = int(input("Faktoriyelini hesaplamak istediğiniz sayıyı girin: "))
    fakt = faktoriyel(sayi)
    print(f"{sayi}! = {fakt}")

    # Personel sınıfı kullanımı
    personel = Personel(1, "Ahmet", "Yılmaz", 1985, "Satış", 25, 40)
    personel.bilgileri_yazdir()

    # Çalışan ve Müdür sınıfı kullanımı
    calisan1 = Calisan("Ayşe", "Demir", 3000)
    calisan2 = Mudur("Mehmet", "Kara", 6000, "Üretim", "12345")

    calisan1.calis()
    calisan2.calis()
```
